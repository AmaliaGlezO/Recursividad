```python
# Descomponer en sumandos
def descomponer(n):
    if n == 0:
        return n
    if n < 0:
        return False
    else:
        for i in range(n):
            for j in range(n):
                if i + j == n:
                    print(f"{i} + {j}")

n = 5
i = n
print(descomponer(n))
```
