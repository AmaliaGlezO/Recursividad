```python
# Sudoku

# Funcin para verificar que en cada filas, columnas y submatriz 3x3estn los nmeros del 1 al 9
def valido(sudoku,n,i,j):
    fila = sudoku[i]
    columna =(f[j] for f in sudoku)
    bloque = [sudoku[a][b]
            for a in range(9)
            for b in range(9)
            if i // 3 == a // 3
            and j // 3 == b // 3]
    print("Es valido el sudoku")
    return n not in fila and n not in columna and n not in bloque

# Funcin para resolver el sudoku
def resolver(sudoku):
    for i in range(9):
        for j in range(9):
            if sudoku[i][j] == 0: 
                for n in range(1, 10):
                    if valido(sudoku, n, i, j):
                        sudoku[i][j] = n
                        if resolver(sudoku):
                            return True
                        else:
                            sudoku[i][j] = 0
                return False 
    print(sudoku)
    return True 

# Funcin para que me imprima el sudoku
print(resolver(sudoku))

sudoku = [
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9],
]    
```
